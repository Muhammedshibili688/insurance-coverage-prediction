```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid", palette="viridis")
plt.style.use("seaborn-v0_8")
sns.set(rc={"figure.figsize": (10,6)})
pd.set_option("display.max_columns", None)
```

```python

import os
os.chdir(r"C:\VScode\Insurance Coverage Prediction")
os.getcwd()

raw = pd.read_csv("data/raw/train.csv")
```

```python
print("\n Dataset Shape:", raw.shape)
print("\n Columns:", list(raw.columns))

print("\n Summary Stats:")
print(raw.describe().T)

print("\n Missing Values:")
print(raw.isna().sum())
```

```python
print("Shape:", raw.shape)
display(raw.head())
display(raw.describe().T)
print("\nMissing values:\n", raw.isna().sum())
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x="shopping_pt", data=raw)
plt.title("Distribution of Shopping Points")
plt.savefig("analysis/visuals/shopping_pt_dist.png", bbox_inches='tight')
plt.show()
```

Most customers dont shop around much

-> Majority stop after 14 shopping points.

-> Very few continue till 1012 quotes.

-> People decide early  quick conversions.


```python
fig, axes = plt.subplots(1,3, figsize=(16,8))

sns.countplot(y="state", data=raw, ax=axes[0], order=raw["state"].value_counts().index[:20])
axes[0].set_title("Top 20 States")

sns.countplot(x="group_size", data=raw, ax=axes[1])
axes[1].set_title("Group Size Distribution")

sns.countplot(x="car_value", data=raw, ax=axes[2], order=raw["car_value"].value_counts().index)
axes[2].set_title("Car Value Categories")

plt.tight_layout()
plt.savefig("analysis/visuals/categorical_analysis.png", bbox_inches='tight')
```

State and location matter

-> Encoded state/location + hashed interactions appear in top feature list.

-> Indicates regional pricing & behavior patterns.

-> Useful for targeted marketing or localized pricing.
